```python
from ns_cmab.environments.causal_bandit_env import SCM, Mechanism,  CausalBanditEnv
```

```python
graph = {
    "yield": ["nutrition", "moisture", "sunlight"]
    ,"nutrition": []
    ,"moisture": []
    ,"sunlight": []
    }
```

```python
f_yield = Mechanism(seed=42)

f_nutr = Mechanism(
    lambda parents, exogenous: exogenous 
    , seed=43
)

f_moist = Mechanism(
    lambda parents, exogenous: exogenous 
    , seed=44
)

f_sun = Mechanism(
    lambda parents, exogenous: exogenous 
    , seed=45
)

mechanisms = {
    "yield": f_yield,
    "nutrition": f_nutr,
    "moisture": f_moist,
    "sunlight": f_sun
}
```

```python
scm = SCM(graph=graph, mechanisms=mechanisms)
```

```python
env = CausalBanditEnv(scm=scm, target_variable="yield")
```
